```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
```

```python
X=df[['Weight','Volume']]
y=df['CO2']
```

```python
regr=linear_model.LinearRegression()
regr.fit(X,y)
```

```python
preCO2=regr.predict([[1500,1200]])
print(preCO2)
```
